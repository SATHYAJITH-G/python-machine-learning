```python
import pandas as pd
import numpy as np
df=pd.DataFrame({'Age':[42,15,67,55,1,29,75,89,410,15,38,22,77]})
print("Before:\n")
print(df)
df['Label']=pd.cut(x=df['Age'],bins=[0,3,17,63,99],labels=['Baby/Toddler','Child','Adult','Elderly'])
print("After:\n")
print(df)
print("Categories:\n")
print(df['Label'].value_counts())
```

```python
import pandas as pd
import numpy as np

# Create DataFrame
df = pd.DataFrame({
    'Age': [150.4,157.6,170,176,164.2,155,159.2,175,162.4,176,153,170.9]
})

print("Before:")
print(df)

# Apply binning on correct column
df['Label'] = pd.cut(
    x=df['Age'],   
    bins=[150,157,169,180],
    labels=['Short','Average','Tall']
)

print("\nAfter:")
print(df)

print("\nCategories:")
print(df['Label'].value_counts())

```
